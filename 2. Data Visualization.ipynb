# Bivariate Analysis 

## Numerical-Categorical

```python
import pandas as pd 
import seaborn as sns 
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
sns.barplot(x="class", y="age", data=df)
```

```python
sns.barplot(x="class", y="fare", data=df, estimator = "sum")
```

```python
sns.barplot(x="class", y="age", data=df, hue="sex")
```

```python
sns.barplot(x="class", y="fare",hue="sex", data=df, estimator = "sum")
```

```python
sns.boxplot(x="class",y="age",hue="survived", data=df)
```

```python
sns.violinplot(x="class",y="age",data=df)
```

```python
sns.distplot(df[df["class"]=="Second"]["age"])
```

## Categorical Data 

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
sns.countplot(x="survived",hue="class", data=df)
```

```python
x = pd.crosstab(df["survived"], df["class"])
x
```

```python
sns.heatmap(x, annot=True, fmt=".1f", cmap="coolwarm")
```

```python
sns.barplot(x="class", y="age", hue="sex",data=df)
```

```python
sns.stripplot(x="class", y="age", hue="sex",data=df)
```

## Numerical

```python
import pandas as pd 
import seaborn as sns 
```

```python
df = sns.load_dataset("titanic")
df.head()
```

### Scatterplot

```python
sns.scatterplot(x="age",y="fare",hue="survived",data=df)
```

```python
df2=sns.load_dataset("tips")
df2.head()
```

```python
sns.scatterplot(x="total_bill",y="tip",data=df2)
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="sex",data=df2)
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="sex",style="smoker",data=df2)
```

### Joint Plot

```python
sns.jointplot(x="total_bill",y="tip",data=df2)
```

```python
sns.jointplot(x="total_bill",y="tip",hue="sex",data=df2)
```

```python
sns.jointplot(x="total_bill",y="tip",kind="reg",data=df2)
```
