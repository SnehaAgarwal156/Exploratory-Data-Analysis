# Outliers

An outlier is an observation in a given dataset that lies far from the rest of the observations. 

```python
sample = [15,101,18,7,17,13,16,11,21,5,15,10,9]
#101 is an outlier
```

Reason for an outlier exists in the dataset

1. Data entry level
2. Variability in the dataset 

   the range of an outliers depends on buisness problems and can change 
   from case to case 

How to treat an outlier? 
1. Trimming
2. Capping 

Various ways of finding an outlier

1. using z score 
2. Using the interquartile range 
3. using percentile 

## Outlier Removal using z-score 

```python
import pandas as pd 
import seaborn as sns 
import numpy as np 
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('scholarship.csv')
data
```

```python
df=data.copy()
df.head()
```

```python
df.shape
```

```python
plt.subplot(1,2,1)
sns.distplot(df['semester_percentage'])
plt.subplot(1,2,2)
sns.distplot(df['scholarship_exam_marks'])
plt.show()
```

```python
mean = df['semester_percentage'].mean()
std=df['semester_percentage'].std()
```

```python
print(mean, ",", std)
```

```python
highest_allowed = mean + 3*std
lowest_allowed = mean - 3*std
```

```python
print(highest_allowed, ",", lowest_allowed)
```

```python
df[(df["semester_percentage"]< lowest_allowed) | (df["semester_percentage"]> highest_allowed)]
```

## Trimming

```python
df[(df["semester_percentage"]> lowest_allowed) & (df["semester_percentage"]< highest_allowed)]
```

```python
df["z_score"] = (df["semester_percentage"]-df["semester_percentage"].mean())/df["semester_percentage"]
```

```python
df.head()
```

```python
df[(df["z_score"]<-3) | (df["z_score"]>3)]
```

```python
df[(df["z_score"]>-3) & (df["z_score"]<3)]
```

## Capping

```python
upper_limit = mean + 3*std
lower_limit = mean - 3*std
```

```python
df["semester_percentage"] = np.where(df["semester_percentage"]> upper_limit, upper_limit, 
                                     np.where(df["semester_percentage"]<lower_limit, lower_limit, df["semester_percentage"]))
```

```python
df[(df["semester_percentage"]< lowest_allowed) & (df["semester_percentage"]> highest_allowed)]
```
