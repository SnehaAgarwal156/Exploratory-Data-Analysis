##  Filling the missing value - Imputation 

i) Numerical data - use mean 

```python
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
df.iloc[:,13:14].values
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(df.iloc[:,13:14].values)
df.iloc[:,13:14] = imputer.transform(df.iloc[:,13:14].values)
```

```python
df
```

```python
# Other method 

df3=df.copy()
mean_value = df3["reviews_per_month"].mean()
df3["reviews_per_month"].fillna(mean_value, inplace=True)
df3.head()
```

```python
df.isna().sum()
```

## Filling missing values - Imputation 

ii) Categorical Data - use most frequent (mode)

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[:,12:13].values
```

```python
impute = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
impute.fit(df.iloc[:,12:13].values)
df.iloc[:,12:13] = impute.transform(df.iloc[:,12:13].values)
```

```python
df.head()
```

```python
# Other method 

df3.nunique()
```

```python
df3['last_review'].fillna(df3["last_review"].value_counts().index[0],inplace=True)
df3.isna().sum()
```

```python
df3.head()
```

```python
df3["name"].fillna("not_filled", inplace=True)
df3.sample(10)
```
