```python
import pandas as pd 
import numpy as np 
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv("googleplaystore.csv")
df.head()
```

```python
impute = SimpleImputer(missing_values = np.nan, strategy='mean')
impute.fit(df.iloc[:,2:3].values)
df.iloc[:,2:3] = impute.transform(df.iloc[:,2:3].values)

df=df.dropna()
df
```

## Ques. Name the apps that are in ART_AND_DESIGN with rating more than 4.5 in ascending order wrt Ratings? 

```python
df_pr=df[df['Category']=='ART_AND_DESIGN']
df_pr=df_pr[df_pr['Rating']>4.5]

df_pr.sort_values(by='Rating', ascending = True)
```

## Ques. Name the apps that are in ART_AND_DESIGN in descending order wrt Reviews? 

```python
df_pr=df[df['Category']=='ART_AND_DESIGN']

df_pr.sort_values(by='Reviews', ascending=False)
```
