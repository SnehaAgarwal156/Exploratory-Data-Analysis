```python
import pandas as pd 
import numpy as np 
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv("googleplaystore.csv")
df.head()
```

```python
impute = SimpleImputer(missing_values = np.nan, strategy='mean')
impute.fit(df.iloc[:,2:3].values)
df.iloc[:,2:3] = impute.transform(df.iloc[:,2:3].values)

df=df.dropna()
df
```

```python
df[df['Category']=='ART_AND_DESIGN']
```

## Ques. How many apps are there in ART_AND_DESIGN with rating more than 4.5?

```python
df_pr=df[df['Category']=='ART_AND_DESIGN']
print(len(df_pr[df_pr['Rating']>4.5]))
```
